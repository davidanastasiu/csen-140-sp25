Define a function that takes an argument `n` and returns a list of odd numbers from [0..n].

```python
def odd_numbers(n):
    result =[]
    # TODO: Implement this function
    for i in range(n + 1):
      if i % 2 != 0:
        result.append(i)
    return result
```

### Problem 2: Read and Sort CSV File
Define a function to read 'students.csv' and return a sorted list.

```python
def read_and_sort_csv(filename):
  with open(filename, 'r') as file:
        lines = file.readlines()
        students = [line.strip().split(',') for line in lines]
        students.sort(key=lambda x: x[0])  # Sort by name
        return students
```

```python
read_and_sort_csv(('students.csv'))
```

### Problem 3: Most Frequent Character
Define a function that counts character occurrences and returns the most frequent one.

```python
def most_frequent_char(s):
    # TODO: Implement this function
    from collections import Counter
    s = s.replace(" ", "")  # Ignore spaces
    count = Counter(s)
    return count.most_common(1)[0][0]
```

```python
most_frequent_char('abbbbc')
```

```python
A = np.random.rand(100, 100)
B = np.random.rand(100, 100)

def add_matrices_loop(A, B):
    # TODO: Implement this function
    rows, cols = A.shape
    result = np.zeros((rows, cols))
    for i in range(rows):
        for j in range(cols):
            result[i][j] = A[i][j] + B[i][j]
    return result
```

### Problem 6: Timing the Methods

```python
import time

def time_execution(func, *args):
    start = time.time()
    result = func(*args)
    end = time.time()
    return end - start

# TODO: Time each function and store results
times_loop = []
times_numpy = []
for i in range(1000):
    t1 = time_execution(add_matrices_loop, A, B)
    times_loop.append(t1)

    t2 = time_execution(add_matrices_numpy, A, B)
    times_numpy.append(t2)
```

### Problem 7: Histogram of Running Times

```python
import matplotlib.pyplot as plt

def plot_histogram(times, method):
    plt.hist(times, bins=30, alpha=0.7, label=method)
    plt.xlabel('Time (seconds)')
    plt.ylabel('Frequency')
    plt.title('Histogram of Execution Times')
    plt.legend()
    plt.show()

# TODO: Collect and plot timing data
plot_histogram(times_loop, "Loop")
plot_histogram(times_numpy, "NumPy")
```

### Problem 8: Your Own Example
Create a simple Python program of your choice in Jupyter Notebook.

```python
def find_prime_numbers_smaller_than(x):
    # find all prime numbers smaller than x
    primes = []
    for num in range(2, x):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    return primes
print(find_prime_numbers_smaller_than(50))
```
